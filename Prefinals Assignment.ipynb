# Undirected Graphs

```python
import matplotlib.pyplot as plt  # Importing matplotlib for visualization
import networkx as nx  # Importing NetworkX for graph manipulation

seed = 55  # Seed for reproducibility
n = 4  # Number of nodes
m = 2  # Number of edges to attach from a new node to existing nodes

# Generate a random graph with n nodes and m edges
G = nx.gnm_random_graph(n, m, seed=seed)

# Add nodes to the graph
G.add_nodes_from([0, 1, 2, 3])

# Define a list of edges to be added to the graph
edgelist = [(0, 1), (1, 2), (2, 3), (0, 3), (2, 2)]

# Add the edges to the graph
G.add_edges_from(edgelist)

# Define the layout of the graph using the spring layout algorithm
pos = nx.spring_layout(G)

# Draw the graph with node labels
nx.draw(G, pos=pos, with_labels=True)

# Display the graph
plt.show()
```

### An undirected graph has two main parts: a set of vertices (points) and a set of edges (lines connecting points). Unlike directed graphs, the edges in an undirected graph don't have a direction; they simply connect two vertices without specifying a starting or ending point.

# Directed Graphs

```python
import matplotlib.pyplot as plt
import networkx as nx
seed = 55
# Use seed for reproducibility
G = nx.DiGraph()
# Add nodes
G.add_nodes_from([0,1, 2, 3])
edgelist = [(0, 1), (1,2), (2,3), (0,3)]
G.add_edges_from(edgelist)
#pos = nx.spring_layout(G, seed=seed)  # Seed for reproducible layout
#pos = nx.planar_layout(G)
pos = nx.circular_layout(G)
nx.draw(G, pos=pos,with_labels=True)
plt.show()
```

### A directed graph, or digraph, has two main parts: a set of vertices (points) and a set of directed edges (arrows between points). Each arrow connects two vertices and points from one to the other.

# Simple Graphs

```python
import matplotlib.pyplot as plt
import networkx as nx
seed = 55
n = 4
m = 2
# Use seed for reproducibility
G = nx.gnm_random_graph(n, m, seed=seed)
# Add nodes
G.add_nodes_from([0,1, 2, 3, 4])
edgelist = [(0, 1), (1,2), (2,3), (0,3)]
G.add_edges_from(edgelist)
#pos = nx.spring_layout(G, seed=seed)
pos = nx.planar_layout(G)
nx.draw(G, pos=pos,with_labels=True)
plt.show()
```

### A simple graph is a type of graph where there are no loops and no multiple edges between the same pair of vertices. In other words, each edge connects two different vertices, and there's only one edge between any pair of vertices.

```python
import matplotlib.pyplot as plt
import networkx as nx
seed = 55
# Use seed for reproducibility
G = nx.gnm_random_graph(n, m, seed=seed)
# Add nodes
G.add_nodes_from([0,1, 2, 3])
edgelist = [(0, 1), (0,2), (0,3), (1,2), (1,3), (2,3)]
G.add_edges_from(edgelist)
#pos = nx.spring_layout(G, seed=seed)  # Seed for reproducible layout
#pos = nx.planar_layout(G)
pos = nx.circular_layout(G)
nx.draw(G, pos=pos,with_labels=True)
plt.show()
```

### A complete graph is a type of graph where every pair of distinct vertices is connected by a unique edge. In simpler terms, in a complete graph, every point (vertex) is directly connected to every other point by a line (edge).

```python
import matplotlib.pyplot as plt
import networkx as nx


G = nx.Graph()
# Add nodes
G.add_nodes_from([0,1, 2, 3], bipartite = 0)
G.add_nodes_from(['A', 'B', 'C', 'D'], bipartite = 1)
edgelist = [(0,'A'), ('A',1), (1,'B'), ('B', 2), (2,'C'), ('C', 3), (3, 'D'), ('C', 0), ('D', 1)]
G.add_edges_from(edgelist)
top = nx.bipartite.sets(G)[0]
pos = nx.bipartite_layout(G, top)
nx.draw(G, pos=pos,with_labels=True)
plt.show()
```

### A bipartite graph is a type of graph whose vertices can be divided into two disjoint sets such that no two vertices within the same set are connected by an edge. In simpler terms, you can split the vertices into two groups, and within each group, there are no direct connections; all connections are between the two different groups.

```python
import matplotlib.pyplot as plt
import networkx as nx

G = nx.Graph()
# Add nodes
G.add_nodes_from([0,1, 2, 3, 4])
edgelist = [(0, 1), (0,2), (0,3), (0,4), (1,2), (4,3)]
G.add_edges_from(edgelist)
pos = nx.spring_layout(G)
nx.draw(G, pos=pos,with_labels=True)
plt.show()
```

### An Euler's circuit is a path in a graph that starts and ends at the same vertex and uses every edge exactly once. In simpler terms, it's a route that travels along every edge in the graph without repeating any edge and ends at the starting point.

```python
import matplotlib.pyplot as plt
import networkx as nx

G = nx.Graph()
# Add nodes
G.add_nodes_from([0,1, 2, 3])
edgelist = [(0, 1), (0,2), (0,3), (1,2)]
G.add_edges_from(edgelist)
pos = nx.shell_layout(G)
nx.draw(G, pos=pos,with_labels=True)
plt.show()
```

### An Euler's trail is a path in a graph that uses every edge exactly once but may not start and end at the same vertex. In simpler terms, it's a route that covers all the edges in the graph once without repeating any edge, but it doesn't necessarily start and finish at the same point.

```python
import matplotlib.pyplot as plt
import networkx as nx
G = nx.Graph()
G.add_nodes_from([0,1, 2, 3, 4])
edgelist = [(0, 1), (0,2), (0,3), (1,2), (1,3), (2,3), (0, 4), (1,4), (2,4), (3,4)]
G.add_edges_from(edgelist)

pos = nx.spring_layout(G)
nx.draw(G, pos=pos,with_labels=True)
plt.show()
```

### A Hamiltonian circuit is a path in a graph that visits every vertex exactly once and ends at the starting vertex, forming a closed loop. In simpler terms, it's a route that travels through every point in the graph once and returns to the starting point without skipping any vertex.
