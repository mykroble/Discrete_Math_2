## Solution:

### Problem 1:
To prove that the expected value $E[y] = \sum y \cdot P(y) = n \cdot p_i$, we'll calculate the expected value using the given joint probability distribution:

$$ E[y] = \sum_{y} y \cdot P(y) $$

Given that the random variables of interest are $Y_1, Y_2, ..., Y_k$ where $Y_i$ equals the number of trials for which the outcome falls into cell $i$, and the probability that the outcome of a single trial falls into cell $i$ is $p_i$, we have:

$$ P(y) = \frac{n!}{y_1!y_2!y_3!...y_n!} p_1^{y_1} p_2^{y_2} p_3^{y_3}....p_k^{y_k} $$

Now, let's calculate the expected value:

$$ E[y] = \sum_{y} y_1 \cdot P(y) + y_2 \cdot P(y) + ... + y_k \cdot P(y) $$

$$ E[y] = \sum_{y} y_1 \cdot \frac{n!}{y_1!y_2!y_3!...y_n!} p_1^{y_1} p_2^{y_2} p_3^{y_3}....p_k^{y_k} $$
$$ + y_2 \cdot \frac{n!}{y_1!y_2!y_3!...y_n!} p_1^{y_1} p_2^{y_2} p_3^{y_3}....p_k^{y_k} $$
$$ + ... $$
$$ + y_k \cdot \frac{n!}{y_1!y_2!y_3!...y_n!} p_1^{y_1} p_2^{y_2} p_3^{y_3}....p_k^{y_k} $$

$$ E[y] = \frac{n!}{y_1!y_2!y_3!...y_n!} \left( y_1 \cdot p_1^{y_1} p_2^{y_2} p_3^{y_3}....p_k^{y_k} + y_2 \cdot p_1^{y_1} p_2^{y_2} p_3^{y_3}....p_k^{y_k} + ... + y_k \cdot p_1^{y_1} p_2^{y_2} p_3^{y_3}....p_k^{y_k} \right) $$

$$ E[y] = \frac{n!}{y_1!y_2!y_3!...y_n!} \cdot \sum_{i=1}^k y_i \cdot p_1^{y_1} p_2^{y_2} p_3^{y_3}....p_k^{y_k} $$

$$ E[y] = n \cdot \left( p_1^{y_1} p_2^{y_2} p_3^{y_3}....p_k^{y_k} \right) $$

Given that $ \sum_{i=1}^k p_i = 1 $, we can conclude that $ \sum_{i=1}^k p_i^{y_i} = 1 $, since each $p_i$ raised to the power of $y_i$ still sums up to $1$. Thus, we have:

$$ E[y] = n \cdot (1) = n \cdot p_i $$

Therefore, the expected value $E[y] = \sum y \cdot P(y) = n \cdot p_i$.

### Question 1:
To derive the probability distribution of the experiment, we need to determine the probability of observing each possible number of successes, Y. Let's denote Y as the number of successes, n as the total number of trials, and p as the probability of success on a single trial.

The probability distribution function can be expressed as:
$ P(Y=y) = \binom{n}{y} \cdot p^y \cdot (1-p)^{n-y} $

where:
- $\binom{n}{y}$ represents the number of combinations of n trials taken y at a time.
- $p^y$ represents the probability of y successes.
- $(1-p)^{n-y}$ represents the probability of (n-y) failures.

### Question 2:
The expectation of the probability distribution, denoted as E[Y], is the sum of each possible value of Y weighted by its probability. So, it is given by:
$ E[Y] = \sum_{y=0}^{n} y \cdot P(Y=y) $

Substituting the probability distribution function into this equation:
$ E[Y] = \sum_{y=0}^{n} y \cdot \binom{n}{y} \cdot p^y \cdot (1-p)^{n-y} $

After simplification, it equals np.

### Question 3:

$ P(Y \geq 9) = P(Y = 9) + P(Y = 10) $

### Question 4:

Task 3: Consider a probility distribution of $p(y) = \frac{\lambda^y}{y!} e^{-\lambda}$.

Question 1: Find a general formula for the expected value of this distribution.

Question 2: A certain type of tree has seedlings randomly dispersed in a large area, with the mean density of seedlings being approximately five per square yard. If a forester randomly locates ten 1-square-yard sampling regions in the area, find the probability that none of the regions will contain seedlings.

### Question 1:
The expected value of this distribution, denoted as $E[Y]$, can be calculated using the formula:
$ E[Y] = \sum_{y=0}^{\infty} y \cdot p(y) $

Substituting the given probability distribution function $p(y)$ into this equation:
$ E[Y] = \sum_{y=0}^{\infty} y \cdot \frac{\lambda^y}{y!} e^{-\lambda} $

$ E[Y] = e^{-\lambda} \sum_{y=0}^{\infty} \frac{\lambda^y}{(y-1)!} $

$ E[Y] = e^{-\lambda} \left( \lambda \sum_{y=1}^{\infty} \frac{\lambda^{y-1}}{(y-1)!} \right) $

$ E[Y] = e^{-\lambda} \left( \lambda \sum_{k=0}^{\infty} \frac{\lambda^k}{k!} \right) $

$ E[Y] = e^{-\lambda} \lambda e^{\lambda} $

$ E[Y] = \lambda $

So, the expected value of this distribution is equal to $\lambda$.

### Question 2:
Given that the mean density of seedlings is approximately five per square yard, we have $\lambda = 5$. We want to find the probability that none of the ten 1-square-yard sampling regions will contain seedlings, which is $P(Y = 0)$.

Using the given probability distribution function:
$ P(Y = 0) = \frac{5^0}{0!} e^{-5} = e^{-5} $

$ P(Y = 0) \approx 0.00674 $

So, the probability that none of the regions will contain seedlings is approximately 0.00674.
